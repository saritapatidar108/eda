```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data1=pd.read_csv("amazon.csv")
data1.dtypes
data1
```

```python
data1.columns
data2=data1
data2
```

```python
data2.info()
```

```python
data2['discounted_price'] = (
    data2['discounted_price']
    .astype(str)
    .str.replace('₹', '', regex=True)  
    .str.replace(',', '', regex=True) 
    .astype(float) 
)
l=data2["discounted_price"]
print(l)
```

```python
data2['actual_price'] = (
    data2['actual_price']
    .astype(str)
    .str.replace('₹', '', regex=True)  
    .str.replace(',', '', regex=True) 
    .astype(float) 
)
k=data2['actual_price']
print(k)
```

```python
data2['discount_percentage'] = (
    data2['discount_percentage']
    .astype(str)
    .str.replace('%', '', regex=True)  
    .astype(float) 
)
s=data2['discount_percentage']
print(s)
```

```python
data2['rating_count'] = (
    data2['rating_count']
    .astype(str)
    .str.replace(',', '', regex=True)  
    .astype(float) 
)
y=data2['rating_count']
print(y)
```

```python
data2['rating'] = pd.to_numeric(data2['rating'], errors='coerce')
k=data2["rating"]
print(k)
```

```python
data2["rating"].isnull().sum()
```

```python
data2.dropna(subset=["rating"],axis=0,inplace=True)
```

```python
#How do ratings differ between products with high and low review counts?
result = data2.groupby('product_name').agg(
    average_rating=('rating',"mean"),
    review_count=('review_content', 'count')
).reset_index()
r1=result.sort_values("review_count",ascending=False).head(1)
rating1=r1['average_rating'].iloc[0] 
print(rating1)
r2=result.sort_values("review_count",ascending=True).head(1)
rating2=r2['average_rating'].iloc[0] 
print(rating2)
diff=rating2-rating1
print(diff)
```
